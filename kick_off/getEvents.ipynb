```python
import imp
from cal_setup import get_calendar_service
from datetime import datetime

service = get_calendar_service()
# Call the Calendar API
now = datetime.utcnow().isoformat() + 'Z' # 'Z' indicates UTC time
print('Getting List of events')
events_result = service.events().list(
    calendarId='primary', timeMin=now,
    singleEvents=True,
    orderBy='startTime').execute()
events_result 
```

```python
events_result['items'][0].keys()
```

```python
bdays = []
# iterate over events, each event being a dictionary
for index, item in enumerate(events_result['items']):
    event_name = item['summary'] # get event name
    event_date = item['start'] # get event date 
    try:
        event_date = event_date['date']
    except:
        event_date = event_date['dateTime'][0:10]
    # to date dtype
    eventDate = datetime.strptime(event_date,'%Y-%m-%d').date()
    if eventDate.year == 2022:
        if "niver" in event_name: # select birthdays only
            if "|" in event_name: # replace | by -
                event_name = event_name.replace("|", "-")
            
            
            try: #try to get Bday's Person Name
                BdayName = event_name.split('-')[1]
                bdays.append({"name" : BdayName,
                                "date" : eventDate})
            except:
                BdayName = event_name.split(' ')[1]
                bdays.append({"name" : BdayName,
                                "date" : eventDate})


for bday in bdays:
    if datetime.now().date() == bday['date']:
        # send message function
        pass
    else:
        print(bday['name'])
```
